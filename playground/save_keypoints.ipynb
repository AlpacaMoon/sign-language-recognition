```python
def save1():
    x = []
    for i in range(trainingCount):
        x.append(inputData[i])
    np.save('new_test.npy', np.array(x))

def save2():
    for i in range(trainingCount):
        np.save(f'testDataFolder/{i}.npy', inputData[i])

from timeit import timeit

testIter = 10
print(timeit(save1, number=testIter))
print(timeit(save2, number=testIter))
```
